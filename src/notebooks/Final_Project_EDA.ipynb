### Subtask:
Combine the relevant columns (`PTS`, `REB`, `AST`, `STL`, `BLK`) from `pre_team_stats_df` and `playoff_team_stats_df` into a single DataFrame. Add a new column, 'Season_Type', to distinguish between 'Pre-Season' and 'Playoff' statistics. This will facilitate side-by-side comparisons.
